### **Q5. How are DataFrame.size() and DataFrame.shape() different?**

### ***ANSWER :***

The `DataFrame.size` and `DataFrame.shape` are both attributes of a Pandas DataFrame that provide information about the dimensions of the DataFrame, but they have different meanings.


>  `DataFrame.size` gives you the total number of elements in the DataFrame ***(rows x columns)***, whereas `DataFrame.shape` provides the dimensions of the DataFrame as a tuple ***(number of rows, number of columns)***.

1. `DataFrame.size`:
   - `DataFrame.size` returns the total number of elements in the DataFrame, which is equivalent to the total number of cells in the DataFrame.
   - The size is calculated by multiplying the number of rows (`DataFrame.shape[0]`) by the number of columns (`DataFrame.shape[1]`).
   - The `size` attribute returns a single integer value representing the total number of elements in the DataFrame.
   - Example: `df.size` returns the total number of elements in DataFrame `df`.

2. `DataFrame.shape`:
   - `DataFrame.shape` returns a tuple that represents the dimensions of the DataFrame.
   - The shape tuple consists of two values: the number of rows and the number of columns, respectively.
   - The first element of `DataFrame.shape` (`DataFrame.shape[0]`) represents the number of rows.
   - The second element of `DataFrame.shape` (`DataFrame.shape[1]`) represents the number of columns.
   - Example: `df.shape` returns the dimensions of DataFrame `df` as a tuple, e.g., `(5, 3)` means the DataFrame has 5 rows and 3 columns.

### Q7. You have a Pandas DataFrame df that contains a column named 'Email' that contains emailaddresses in the format 'username@domain.com'. Write a Python function that creates a new column 'Username' in df that contains only the username part of each email address.
### The username is the part of the email address that appears before the '@' symbol. For example, if the email address is 'john.doe@example.com', the 'Username' column should contain 'john.doe'. Your function should extract the username from each email address and store it in the new 'Username'column.


### ***ANSWER :***

````python
import pandas as pd

data = {'Email': ["john.doe@example.com","user1@gmail.com","user2@gmail.com","user3@gmail.com"]}

df = pd.DataFrame(data)
#Create a new column "Username" which extracts the name from the email
df['Username'] = df['Email'].apply(lambda x: x.split("@")[0])

df
````

### **Q8. You have a Pandas DataFrame df with columns 'A', 'B', and 'C'. Write a Python function that selects all rows where the value in column 'A' is greater than 5 and the value in column 'B' is less than 10. The function should return a new DataFrame that contains only the selected rows.**

### **For example, if df contains the following values:**

```
   A B C

0  3 5 1

1  8 2 7

2  6 9 4

3  2 3 5

4  9 1 2

```
### **Your function should select the following rows:**

```
  A B C
1 8 2 7
4 9 1 2
```
### **The function should return a new DataFrame that contains only the selected rows.**


### ***ANSWER :***


````python
import pandas as pd
#Create a DataFrame
data = {'A' : [3,8,6,2,9],
        'B' : [5,2,9,3,1],
        'C' : [1,7,4,5,2]}


df = pd.DataFrame(data)
````

````python
df
````

````python

def selects_all_rows(df):

  #value in column 'A' is greater than 5 and the value in column 'B' is less than 10.
  selected_rows = df[(df['A'] > 5) & (df['B'] < 10)]
  return selected_rows
````

````python
selected_rows_df = selects_all_rows(df)
#new DataFrame that contains only the selected rows.
selected_rows_df
````

### **Q9. Given a Pandas DataFrame df with a column 'Values', write a Python function to calculate the mean, median, and standard deviation of the values in the 'Values' column.**

### ***ANSWER :***

````python
def stats(df):
  mean = df['Values'].mean()
  median = df['Values'].median()
  standard_deviation = df['Values'].std()

  stats_data = {'Mean': [mean] ,
                'Median' : [median] ,
                'Standard_Deviation' : [standard_deviation]
                }

  return stats_data
````

````python
df = pd.DataFrame({'Values': [10,20,30,40]})
stat_df = stats(df)

df2 = pd.Series(stat_df)
df2
````

```
    Date

3 2023-01-04
```

```
2 2023-01-03    Tuesday
3 2023-01-04    Wednesday
4 2023-01-05    Thursday
```

### **Q12. Given a Pandas DataFrame df with a column 'Date' that contains timestamps, write a Python function to select all rows where the date is between '2023-01-01' and '2023-01-31'.**


### ***ANSWER :***

````python
import pandas as pd
````

````python

# Assuming df is your DataFrame with the 'Date' column
df = pd.DataFrame({'Date': ['2023-01-01 12:00:00', '2023-01-20 09:30:00', '2023-01-31 16:15:00', '2023-02-10 14:20:00']})

# Convert the 'Date' column to datetime
df['Date'] = pd.to_datetime(df['Date'])

# Define the start and end dates
start_date = pd.to_datetime('2023-01-01')
end_date = pd.to_datetime('2023-01-31')

# Calculate the time duration using pd.Timedelta()
time_duration = pd.Timedelta(days=31)

# Filter the DataFrame based on the date range
selected_rows = df[(df['Date'] >= start_date) & (df['Date'] <= (start_date + time_duration))]

# Print the selected DataFrame with timestamps
print(selected_rows)
````
