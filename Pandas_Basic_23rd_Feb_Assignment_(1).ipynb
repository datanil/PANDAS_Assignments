### **Q4. What is ‘DataFrame’ in pandas and how is it different from pandas.series? Explain with an example.**

***ANSWER :***


*   **Series :**

> Series is a one-dimensional labeled array  in pandas that can hold data of a single data type. It can be seen as a single column in a DataFrame. A DataFrame is composed of one or more Series, where each Series represents a column.


*  **DataFrame :**

> In pandas, a DataFrame is a two-dimensional labeled data structure that can hold data of different types (e.g., numerical, string, boolean, etc.) in a tabular format. It can be thought of as a table with rows and columns, similar to a spreadsheet or a SQL table. Each column in a DataFrame is a Series.


*Here is an example to explain the difference between a DataFrame and Series :*







```python
#Series
import pandas as pd

num_list = [1,2,3,4,5]
series = pd.Series(num_list)
print(series)
```

In the above code, we create a Series named series with five integer values. The output shows the values along with their corresponding index.

```python
from pandas.io.formats.style_render import DataFrame
#Data frame
import pandas as pd
num_data = {
    "Numbers" : [1,2,3,4,5]
}

df = pd.DataFrame(num_data)
print(df)
```

In this example, we create a DataFrame named df with a single column labeled 'Numbers'. The output shows the DataFrame with the values in the 'Numbers' column and their corresponding index. Note that the DataFrame has a tabular structure with column names, while the Series is a one-dimensional object without column names.

### **Q5. What are some common functions you can use to manipulate data in a Pandas DataFrame? Can you give an example of when you might use one of these functions?**


### ***ANSWER :***

Pandas provides a wide range of functions to manipulate data in a DataFrame. Here are some common functions you can use:



*   **head() :**


> his function allows you to view the first five rows of a DataFrame. It is useful to quickly inspect the data and get a sense of its structure.

* ***Example :***





```python
import pandas as pd
data = {"Name" : ["Ram" , "Seema" , "Amar" , "Rohit" , "Kumari" , "Amita" , "Akbar" , "Roshan"],
        "Age" : [20 ,18 ,32 , 25 , 23 , 20 , 19 , 27],
        "Gender" : ["Male" , "Female", "Male" , "Male" , "Female", "Female" , "Male" , "Male"]
        }
df = pd.DataFrame(data)
print(df)

df.head()
```



*   **describe :**

> This function provides summary statistics of numerical columns in the DataFrame, such as count, mean, standard deviation, minimum, quartiles, and maximum. It helps in understanding the distribution of the data.



*   ***Example :***

```python
data = {"Name" : ["Ram" , "Seema" , "Amar" , "Rohit" , "Kumari" , "Amita" , "Akbar" , "Roshan"],
        "Age" : [20 ,18 ,32 , 25 , 23 , 20 , 19 , 27],
        "Gender" : ["Male" , "Female", "Male" , "Male" , "Female", "Female" , "Male" , "Male"]
        }
df = pd.DataFrame(data)
print(df)

df.describe()
```

*   **tail() :**


> his function allows you to view the last five rows of a DataFrame. It is useful to quickly inspect the data and get a sense of its structure.

* ***Example :***

```python
data = {"Name" : ["Ram" , "Seema" , "Amar" , "Rohit" , "Kumari" , "Amita" , "Akbar" , "Roshan"],
        "Age" : [20 ,18 ,32 , 25 , 23 , 20 , 19 , 27],
        "Gender" : ["Male" , "Female", "Male" , "Male" , "Female", "Female" , "Male" , "Male"]
        }
df = pd.DataFrame(data)
print(df)

df.tail()
```

*   **sort_values() :**


> This function allows you to sort the DataFrame based on one or more columns. It is useful when you want to order the data based on specific criteria..

* ***Example :***

```python
data = {"Name" : ["Ram" , "Seema" , "Amar" , "Rohit" , "Kumari" , "Amita" , "Akbar" , "Roshan"],
        "Age" : [20 ,18 ,32 , 25 , 23 , 20 , 19 , 27],
        "Gender" : ["Male" , "Female", "Male" , "Male" , "Female", "Female" , "Male" , "Male"]
        }
df = pd.DataFrame(data)
print(df)

df.sort_values(by='Age')
```

These are just a few examples of the many functions available in Pandas for manipulating data in a DataFrame.

### **Q6. Which of the following is mutable in nature Series, DataFrame, Panel?**

### ***ANSWER :***

* **Series:**

> A pandas Series is mutable, meaning you can
modify its elements by assigning new values to specific indices or slices of the Series. You can change the values, add new elements, or delete existing elements within a Series.


* **DataFrame:**

> Similar to Series, a pandas DataFrame is also mutable. You can modify the values of individual cells, add or remove columns, and update rows within a DataFrame.



* **Panel:**

> The pandas Panel, on the other hand, is not mutable. The Panel structure was deprecated in pandas version 0.20.1, and it is recommended to use alternative data structures like MultiIndex DataFrames or 3D NumPy arrays instead.



### **Q7. Create a DataFrame using multiple Series. Explain with an example.**

### ***ANSWER :***

```python
import pandas as pd
name_series = pd.Series(["Ram" , "Seema" , "Amar" , "Rohit" , "Kumari" , "Amita" , "Akbar" , "Roshan"])
age_series = pd.Series([20 ,18 ,32 , 25 , 23 , 20 , 19 , 27])
gender_series = pd.Series(["Male" , "Female", "Male" , "Male" , "Female", "Female" , "Male" , "Male"])

df = pd.DataFrame({"NAME" : name_series , "AGE" : age_series , "GENDER": gender_series})
df
```

In the above code, we first create three individual Series: name_series, age_series, and gender_series. Each Series represents a column in the desired DataFrame. The name_series contains names, age_series contains ages, and gender_series contains gender information.

Next, we combine these Series into a DataFrame using the pd.DataFrame() function. We pass a dictionary as an argument to the function, where the keys represent column names and the values represent the corresponding Series.

Finally, we print the resulting DataFrame df, which contains the combined data from the individual Series.

This example demonstrates how you can create a DataFrame by combining multiple Series, where each Series represents a column of data.
